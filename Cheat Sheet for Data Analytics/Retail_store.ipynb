```python
import matplotlib.pyplot as plt #the required library


dfts = df5.copy()

dfts = dfts.set_index('InvoiceDate')

monthly_totals = dfts.resample('W').sum('Quantity')

plt.plot(monthly_totals['Quantity'])
plt.xlabel('Week')
plt.ylabel('Total Quantity')
plt.title('Monthly Quantity Totals')
plt.ticklabel_format(style='plain', axis='y')
#plt.tight_layout()
plt.xticks(rotation=60)
plt.show()
```
