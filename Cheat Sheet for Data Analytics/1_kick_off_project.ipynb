7. Is there a difference of average room space (Space divided by Rooms) for apartments built after 2000 versus apartment built before 2000?

```python
appartments_after_2000 = housing_berlin[housing_berlin['Year_Construction'] > 2000]
apparments_before_2000 = housing_berlin[housing_berlin['Year_Construction'] <= 2000]

avg_room_space_after_2000 = appartments_after_2000['Space'] / appartments_after_2000['Rooms']
avg_room_space_before_2000 = apparments_before_2000['Space'] / apparments_before_2000['Rooms']

print(avg_room_space_after_2000.mean())
print(avg_room_space_before_2000.mean())
```

8. What's the percentage of the apartments in high quality condition?

```python
set(housing_berlin.Condition)
```

```python
high_quality_condition = ['first_time_use', 'mint_condition','refurbished','first_time_use_after_refurbishment','fully_renovated']

high_quality_appartments = housing_berlin[housing_berlin['Condition'].isin(high_quality_condition)]

percentage_high_quality = len(high_quality_appartments) / len(high_quality_condition) * 100

print("Percentage of apartments in high-quality condition: {:.2f}%".format(percentage_high_quality))
```
